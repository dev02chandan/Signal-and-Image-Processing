```python
#plotting sine wave

# x=A*sin(2*pi*f*t)

t = np.arange(0,1,0.1)
f = 2
A = 1
x = A*np.sin(2*np.pi*f*t)
y = A*np.cos(2*np.pi*f*t)
```

```python
print(t , "\n\n")
print(x)
```

```python
plt.plot(t,x)
# now the t is only 10 samples 
# so the wave is not smooth
# t = np.arange(0,1,0.01) - this will give a smooth curve
```

```python
plt.plot(t,x ,"g*-")
# "r" gives a red colour plot
#" r* " gives 
# https://matplotlib.org/stable/api/markers_api.html
plt.xlabel("Time ---->")
plt.ylabel("Amplitude--->")
plt.title("Sine Wave")
```

```python
plt.plot(t,x ,"g*-" , t , y , "r*-")
# here we did cos vs sin wave 
# cos = red and sin = green
plt.xlabel("Time ---->")
plt.ylabel("Amplitude--->")
plt.title("Sine Wave")
```

```python
# Plotting things side by side 

plt.figure(figsize=(7,7)) # increasing the size 

#bg colour - can change - check documents 


plt.subplot(1,2,1)
plt.plot(t,x)
plt.subplot(1,2,2)
plt.plot(t,y)

# Subplot
# 1,2 means 1 row and 2 columns
# 1,2, "1" means that this is the first one in the subplot
# 1,2, "2" means that this is the second one in the subplot
```

```python
plt.subplot(3,2,1)
plt.plot(t,x)
plt.subplot(1,2,2)
plt.plot(t,y)
#playing around with it
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.plot(t,x)
plt.subplot(1,3,2)
plt.stem(t,x)
#Stem is used for discrete time signals 
plt.subplot(1,3,3)
plt.plot(t,y)
```
