### ii. Thresholding

```python
img1= cv2.imread('Pollen.tif',0)
plt.imshow(img1, cmap="gray")
plt.show()
print(img1)
```

```python
val = int(input("Enter Threshold Value: "))
```

```python
new_img = img1.copy()
```

```python
for i in range(img1.shape[0]):
  for j in range(img1.shape[1]):
    if img1[i,j] >= val:
      new_img[i, j] = 255

    else:
      new_img[i, j] = 0

new_img

plt.imshow(new_img, cmap='gray')
```

### iii. and iv.Gray level Slicing 

```python
img = cv2.imread("kidney.tif",0)
plt.imshow(img, cmap="gray")
```

```python
upper_limit= int(input("Upper Limit: "))
lower_limit = int(input("Lower Limit: "))
s2 = int(input("s2 value: "))
s1 = int(input("s1 value: "))
```

```python
new_img = np.zeros(img.shape)
```

```python
# without background

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if img[i,j] <= upper_limit and img[i,j] >= lower_limit:
      new_img[i, j] = s2

    else:
      new_img[i, j] = s1

new_img

plt.imshow(new_img, cmap='gray')
```

```python
# With background

for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if img[i,j] <= upper_limit and img[i,j] >= lower_limit:
      new_img[i, j] = s2

new_img

plt.imshow(new_img, cmap='gray')
```

### Blurring

```python
img= cv2.imread('/content/Pollen.tif',0)
m,n=img.shape 
```

```python
mask= np.ones([3,3],dtype=int)
print(mask)
mask=mask/9
```

```python
print(mask)
```

```python
img_new=img.copy()
for i in range(1,m-1):
    for j in range(1,n-1):
        temp= img[i-1,j-1]*mask[0,0]+img[i-1,j]*mask[0,1]+img[i-1,j+1]*mask[0,2]+img[i,j-1]*mask[1,0]+img[i,j]*mask[1,1]+img[i,j+1]*mask[1,2]+img[i+1,j-1]*mask[2,0]+img[i+1,j]*mask[2,1]+img[i+1,j+1]*mask[2,2]
        img_new[i,j]=temp
        
#img_new=img_new.astype(np.uint8)
cv2.imwrite('blurred.png',img_new)
```

```python
plt.imshow(img_new, cmap="gray", vmin=0, vmax=255)
```

```python
img_gen= np.zeros([m,n], dtype=int)
b= int(input("Enter size of the mask: "))
a=b//2

for i in range(a,m-a):
    for j in range(a,n-a):
        temp= img[i-a:i+a+1,j-a:j+a+1] 
        
        constant= np.sum(temp)
        img_gen[i,j]= constant//b**2
```

```python
plt.imshow(img_gen,cmap="gray" )
```
