###Importing

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Grayscale Resolution

```python
#Read the original image and know its type
img1= cv2.imread('/content/lake.tif',0)
#type(img1)
np.amax(img1) 
```

```python
[m,n]=img1.shape
print(m,n)
```

####Taking Value of b = 2

```python
b= int(input("Enter the number of bits for the new image: "))
```

```python
#Calculate the constant value to do the mapping
maxold= 256
maxnew= 2**b
constant = (maxnew-1)/(maxold-1)
constant
```

We can say that the new image will be 0.012 times the size of the original image.


Compression Ratio = 2^8 / 2^2
= 64

```python
img2 = constant*img1

# directly multiplying the image and the constant because image is a numpy arrays
```

```python
img2= np.round(img2)
```

```python
plt.imshow(img1,cmap="gray")
```

```python
plt.imshow(img2, cmap='gray')
cv2.imwrite("grey7.png", img2)
#cv2.imwrite("binary.png",img2)
```

```python
fig = plt.figure(figsize=(10,10),facecolor='y')
plt.subplot(1,2,1)
plt.imshow(img1,cmap="gray")
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(img2,cmap="gray")
plt.title("New image with number of bits={}".format(b))
```
